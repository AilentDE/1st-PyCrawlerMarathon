## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = bs(r.text, 'html5lib')
```

```python
soup
```

### ① 取出今天所有的發文

```python
all_list = soup.find(class_='part_list_2').find_all('h3')
for x in all_list:
    print(x.find(class_='date').text, x.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
data = []
for x in all_list:
    temp = []
    temp.append(x.find(class_='tag').text)
    temp.append(x.find(class_='date').text)
    temp.append(x.find_all('a')[-1].text.replace('\u3000', ' '))
    
    data.append(temp)
```

```python
import pandas as pd
df = pd.DataFrame(data, columns=['tag', 'date', 'title']).sort_values(by=['tag', 'date'])
df[df['tag']=='國際']
```

### ③ 哪一個類別的文章最多

```python
df['tag'].value_counts()
```
