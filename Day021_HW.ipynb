## 1. 取出今天所有的新聞

```python
import requests
from bs4 import BeautifulSoup

#先取出基本的新聞列表
url = 'https://www.ettoday.net/news/news-list-2020-4-6-0.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html5lib')

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
# JS加載的新聞列表
url = 'https://www.ettoday.net/show_roll.php'
payload = {'offset': 1,
           'tPage': 3,
           'tFile': '20200406.xml',
           'tOt': 0,
           'tSi': 100,
           'tAr': 0}
while True:
    r = requests.post(url, data=payload)
    soup = BeautifulSoup(r.text, 'html5lib')
    for d in soup.find_all('h3'):
        if '2020/04/05' in d.find('span').text:
            break
        else:
            print(d.find('span').text, d.find('a').text)
    
    if '2020/04/06' not in soup.find('span').text:
        break
    payload['offset']+=1
```

## 2. 取出現在時間兩小時內的新聞

```python
import time
import pandas as pd
x = time.strftime('%Y/%m/%d %H:%M', time.localtime())#現在時間

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html5lib')

for d in soup.find(class_="part_list_2").find_all('h3'):
    if pd.to_datetime(x) - pd.to_datetime(d.find(class_='date').text) < pd.Timedelta(hours=2):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    else:
        break
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
# from selenium import webdriver
# from selenium.webdriver.support.ui import Select
# import time

browser = webdriver.Chrome()
browser.get("https://www.ettoday.net/news/news-list.htm")

time.sleep(2)

selectSite = Select(browser.find_element_by_id("selD"))
selectSite.select_by_value('4')

browser.find_element_by_id('button').click()
```

```python
# 每個兩秒鐘使用JS移動到網頁最下方：

# import time
for i in range(13):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if (pd.to_datetime(d.find(class_="date").text) <= pd.to_datetime('2020/04/04 17:00')) and\
    (pd.to_datetime(d.find(class_='date').text)) >= pd.to_datetime('2020/04/04 15:00'):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
