## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')

import pandas as pd
import json
data = json.loads(r.text)
a = len(data)
b = pd.DataFrame(data[0].keys(), columns=['Columns'])

print('API回傳{}筆資料。\n所有欄位名單：\n{}'.format(a, b))
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

'''
Your Code
'''
columns = ['title', 'updatedAt', 'commentCount', 'likeCount']
data1 = []
for i in range(len(data)):
    data2 = []
    for j in columns:
        data2.append(data[i][j])
    data1.append(data2)
dataframe = pd.DataFrame(data1, columns=columns)
dataframe
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

'''
Your Code
'''
final_data = []
data_average = []
data_average.append(dataframe['commentCount'].mean())
data_average.append(dataframe['likeCount'].mean())
final_data.append(data_average)

# unpopular data
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
data = json.loads(r.text)
data1 = []
for i in range(len(data)):
    data2 = []
    for j in columns:
        data2.append(data[i][j])
    data1.append(data2)
dataframe = pd.DataFrame(data1, columns=columns)
data_average = []
data_average.append(dataframe['commentCount'].mean())
data_average.append(dataframe['likeCount'].mean())
final_data.append(data_average)

pd.DataFrame(final_data, columns=['ave_Comment', 'ave_Like'], index=['Popular', 'UnPopular'])
```
