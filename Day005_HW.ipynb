```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

'''
Your Code
'''
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')

import pandas as pd
import json
data = json.loads(r.text)
a = len(data)
b = pd.DataFrame(data[0].keys(), columns=['Columns'])

print('APIå›å‚³{}ç­†è³‡æ–™ã€‚\næ‰€æœ‰æ¬„ä½åå–®ï¼š\n{}'.format(a, b))
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

'''
Your Code
'''
columns = ['title', 'updatedAt', 'commentCount', 'likeCount']
data1 = []
for i in range(len(data)):
    data2 = []
    for j in columns:
        data2.append(data[i][j])
    data1.append(data2)
dataframe = pd.DataFrame(data1, columns=columns)
dataframe
```
