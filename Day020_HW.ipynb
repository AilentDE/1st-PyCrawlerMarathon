```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html5lib')
table = soup.find('table', class_='TABLE_G').find('tbody')
table
```

```python
table_list = table.find_all('td')
df = {}

for x in table_list:
    try:
        if x['style'] == 'white-space:nowrap;':
            temp = {}
            ind = table_list.index(x)
            rows = int(x['rowspan'])
            for y in range(rows):
                temp[table_list[ind+2].text] = table_list[ind+3].text
                ind +=3
            
            df[x.text] = temp
    except:
        continue
```
